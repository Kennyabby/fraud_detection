```python
from distance_module import user_transaction_distance, transaction_distance
import matplotlib.pyplot as plt
from scipy.stats import norm
from math import sqrt
```

```python
# The class object created below is a User object, that contains methods and parmeters that allow easy access to the functions
# in the distance_module and statistics_module. 

# Note that the methods of this User class are short numbers, this is done so that the methods are easily accesible.
class User:
    def __init__(self, user_id,transaction_file,user_transactions,amount_list,get_all=False):
        self.user_id = user_id
        self.transaction_file = transaction_file        
        self.user_transactions = user_transactions
        self.amount_list = amount_list
        self.get_all = get_all
    def _0(self,transaction1,transaction2):
        return [{'distance':user_transaction_distance(transaction1, transaction2)}]
    def _1(self,transaction1, transaction2):
        return [{'distance':transaction_distance(transaction1, transaction2)}]
    def _2(self):
        return get_average_user_transaction(self.transaction_file, self.get_all, self.user_id)
    def _3(self):
        return get_user_transactions_mode(self.transaction_file, self.get_all, self.user_id)
    def _4(self):
        return get_user_transactions_median(self.transaction_file, self.get_all, self.user_id)
    def _5(self):
        return get_user_transactions_interquartile_range(self.transaction_file, self.get_all, self.user_id)
    def _6(self):
        return get_location_centroid(self.transaction_file, self.get_all, self.user_id)
    def _7(self):
        return get_transaction_standard_deviation(self.transaction_file, self.get_all, self.user_id)
    def _8(self, test_amount=''):
        return get_user_transaction_zscore(self.transaction_file, self.get_all, self.user_id, test_amount)
    def _9(self, test_x='', test_y=''):
        return get_outliers(self.transaction_file, self.get_all, self.user_id, test_x, test_y)
    def _10(self, test_amount=''):
        return get_user_abnormal_transactions(self.transaction_file, self.get_all, self.user_id, test_amount)
    def _11(self,amount='', x='', y='', threshold=''):
        
        amount_list = self.amount_list
        amount_mean = sum(amount_list)/len(amount_list)
        mean_diff_square_sum=0
        for amt in amount_list:            
            mean_diff = amt - amount_mean            
            mean_diff_square = pow(mean_diff,2)
            mean_diff_square_sum+=mean_diff_square
        
        amount_stdev = sqrt(mean_diff_square_sum/len(amount_list) - 1)
        plt.plot(amount_list, norm.pdf(amount_list, amount_mean, amount_stdev))
        plt.show()
        if amount=='' and x=='' and y=='':
            trans_list =[]
            for trans in self.user_transactions:
                trans_list+=[is_fraudulent(self.transaction_file, self.user_id, trans, amount, x, y, threshold)]
            return trans_list        
        return [is_fraudulent(self.transaction_file, self.user_id, amount=amount, x=x, y=y, threshold=threshold)]
    def _12(self, x='', y=''):
        if x=='' and y=='':
            trans_list =[]
            for trans in self.user_transactions:
                trans_list+=[get_location_transaction_frequency(self.transaction_file, self.user_id, trans, x, y)]
            return trans_list
        return [get_location_transaction_frequency(self.transaction_file, self.user_id,x=x, y=y)]
    def _13(self,n=100):
        return get_transaction_nth_percentile(self.transaction_file, self.get_all, n, self.user_id)
```
